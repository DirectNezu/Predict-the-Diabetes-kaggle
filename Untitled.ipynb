```python
train['Glucose'] = train['Glucose'].replace(0,train['Glucose'].mean())
test['Glucose'] = test['Glucose'].replace(0,test['Glucose'].mean())

train['BloodPressure'] = train['BloodPressure'].replace(0,train['BloodPressure'].mean())
test['BloodPressure'] = test['BloodPressure'].replace(0,test['BloodPressure'].mean())

train['BMI'] = train['BMI'].replace(0, train['BMI'].median())
test['BMI'] = test['BMI'].replace(0, test['BMI'].median())

train['SkinThickness'] = train['SkinThickness'].replace(0, train['SkinThickness'].median())
test['SkinThickness'] = test['SkinThickness'].replace(0, test['SkinThickness'].median())

train['Insulin'] = train['Insulin'].replace(0, train['Insulin'].median())
test['Insulin'] = test['Insulin'].replace(0, test['Insulin'].median())
```

```python
X = train.drop(columns=['Outcome'])
y = train['Outcome']

y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

os = SMOTE(random_state = 1, k_neighbors = 1)

column = X.columns

X_os, y_os = os.fit_resample(X, y)
X_os = pd.DataFrame(data = X_os, columns = column)
y_os = pd.DataFrame(data = y_os, columns = ['Outcome'])

print(X_os.shape, y_os.shape)
y_os.value_counts()
```

```python
X_os.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_os, y_os, random_state = 123, test_size = 0.4) 
print(X_train.shape, X_test.shape)
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score
model1= CatBoostClassifier()

model1.fit(X_train, y_train)

answers_pred=model1.predict(X_test)

print(accuracy_score(y_test,answers_pred))
print(recall_score(y_test,answers_pred))
print(roc_auc_score(y_test,answers_pred))
```

```python
param_grid = { 'n_estimators': range (10, 61, 10),
              'max_depth': range (1,15, 2),
              'min_samples_leaf': range (1,9),
              'min_samples_split': range (2,12,2)}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
clf1 = RandomForestClassifier(random_state=42)
grid = GridSearchCV(clf1,param_grid,cv=5,scoring="accuracy",n_jobs=-1)
grid.fit(X_train,y_train)
```

```python
best_r_params=grid.best_estimator_.get_params()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score
model2 = RandomForestClassifier(**best_r_params)

model2.fit(X_train, y_train)

answers_pred=model2.predict(X_test)

print(accuracy_score(y_test,answers_pred))
print(recall_score(y_test,answers_pred))
print(roc_auc_score(y_test,answers_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score
model3 = LogisticRegression()

model3.fit(X_train, y_train)

answers_pred=model3.predict(X_test)

print(accuracy_score(y_test,answers_pred))
print(recall_score(y_test,answers_pred))
print(roc_auc_score(y_test,answers_pred))
```

```python
from xgboost import XGBClassifier
model4=XGBClassifier(base_score=0.5,max_depth=3,n_estimators=70,n_jobs=10)
model4.fit(X_train,y_train)
answers_pred=model4.predict(X_test)

print(accuracy_score(y_test,answers_pred))
print(recall_score(y_test,answers_pred))
print(roc_auc_score(y_test,answers_pred))
```
